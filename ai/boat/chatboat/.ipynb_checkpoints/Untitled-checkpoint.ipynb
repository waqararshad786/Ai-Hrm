```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import os
import pickle
from PyPDF2 import PdfReader
from termcolor import colored
import warnings
warnings.filterwarnings('ignore')

# -----------------------------
# 1. Load model with fallback
# -----------------------------
def load_model():
    """Try multiple models in case download fails"""
    models_to_try = [
        "all-MiniLM-L6-v2",        # Best balance of speed/accuracy
        "paraphrase-MiniLM-L3-v2",  # Very small, fast
        "all-mpnet-base-v2",        # Original (largest)
    ]
    
    # First check if models exist locally
    local_model_path = "./local_models/all-MiniLM-L6-v2"
    if os.path.exists(local_model_path):
        try:
            print(colored(f"📦 Loading local model...", "yellow"))
            return SentenceTransformer(local_model_path)
        except:
            pass
    
    # Try downloading
    for model_name in models_to_try:
        try:
            print(colored(f"📥 Attempting to load {model_name}...", "yellow"))
            model = SentenceTransformer(model_name)
            print(colored(f"✅ Successfully loaded {model_name}", "green"))
            return model
        except Exception as e:
            print(colored(f"⚠️ Could not load {model_name}: {str(e)[:100]}", "red"))
            continue
    
    print(colored("❌ Could not load any model. Using simple mode.", "red"))
    return None

# Load model
model = load_model()

# -----------------------------
# 2. HR Knowledge Base
# -----------------------------
knowledge_base = []

def extract_text_from_pdf(pdf_path):
    try:
        reader = PdfReader(pdf_path)
        text = []
        for page in reader.pages:
            page_text = page.extract_text()
            if page_text:
                text.append(page_text)
        return " ".join(text)
    except:
        return ""

docs_folder = "../hr_docs"

# Check if folder exists
if not os.path.exists(docs_folder):
    os.makedirs(docs_folder)
    print(colored(f"📁 Created {docs_folder} folder. Please add your HR text files.", "yellow"))

# Load documents
for file in os.listdir(docs_folder):
    file_path = os.path.join(docs_folder, file)
    try:
        if file.endswith(".pdf"):
            text = extract_text_from_pdf(file_path)
            if text:
                paragraphs = text.split("\n\n")
                knowledge_base.extend([p.strip() for p in paragraphs if p.strip()])
        elif file.endswith(".txt"):
            with open(file_path, "r", encoding="utf-8") as f:
                content = f.read()
                # Split by Q: pattern for better chunks
                if "Q:" in content:
                    chunks = content.split("Q:")
                else:
                    chunks = content.split("Q:")
                knowledge_base.extend([c.strip() for c in chunks if c.strip()])
    except Exception as e:
        print(colored(f"Error loading {file}: {e}", "red"))

print(colored(f"📚 Loaded {len(knowledge_base)} HR paragraphs.", "green"))

# -----------------------------
# 3. Encode or load embeddings
# -----------------------------
hr_embeddings = None
if model and knowledge_base:
    embeddings_file = "hr_doc_embeddings.pkl"
    if os.path.exists(embeddings_file):
        try:
            with open(embeddings_file, "rb") as f:
                hr_embeddings = pickle.load(f)
            print(colored("✅ Loaded existing embeddings", "green"))
        except:
            hr_embeddings = model.encode(knowledge_base, show_progress_bar=True)
            with open(embeddings_file, "wb") as f:
                pickle.dump(hr_embeddings, f)
    else:
        hr_embeddings = model.encode(knowledge_base, show_progress_bar=True)
        with open(embeddings_file, "wb") as f:
            pickle.dump(hr_embeddings, f)

# -----------------------------
# 4. Simple responses for fallback
# -----------------------------
simple_responses = {
    "attendance": "Please check the attendance dashboard at portal.company.com/attendance",
    "leave": "You can apply for leave in the Attendance Dashboard under 'Apply Leave'",
    "payroll": "Payroll questions? Contact payroll@company.com",
    "holiday": "Company holidays are listed in the HR portal",
    "benefits": "Check benefits information in the employee portal",
}

def simple_fallback_response(user_input):
    user_input = user_input.lower()
    for key, response in simple_responses.items():
        if key in user_input:
            return colored(response, "cyan")
    return colored("I'm having trouble connecting. Please contact HR directly at hr@company.com", "yellow")

# -----------------------------
# 5. Learning / Advice Module
# -----------------------------
learning_topics = [
    {"topic": "Python", "advice": "Start with basics: variables, loops, and functions. Practice small exercises daily."},
    {"topic": "Data Science", "advice": "Learn numpy, pandas, matplotlib, then do small projects."},
    {"topic": "Machine Learning", "advice": "Understand supervised and unsupervised learning; practice with scikit-learn."},
    {"topic": "HR Study", "advice": "Review HR policies, attendance rules, leave procedures, and payroll regulations."},
]

advice_embeddings = None
if model:
    advice_texts = [t["advice"] for t in learning_topics]
    advice_embeddings = model.encode(advice_texts)

def get_learning_advice(user_input, threshold=0.45, top_n=3):
    if not model or advice_embeddings is None:
        return None
        
    user_emb = model.encode([user_input])
    sims = cosine_similarity(user_emb, advice_embeddings)[0]
    best_idx = np.argmax(sims)
    score = sims[best_idx]

    if score >= threshold:
        return colored(f"📘 Advice: {learning_topics[best_idx]['advice']} (Confidence: {score:.2f})", "cyan")
    else:
        top_indices = sims.argsort()[-top_n:][::-1]
        suggestions = [learning_topics[i]["advice"] for i in top_indices]
        return colored("I'm not fully confident. Here are some advice suggestions:\n- " + "\n- ".join(suggestions), "cyan")

# -----------------------------
# 6. Greeting Module
# -----------------------------
greetings = {
    "hello": "Hello! How can I assist you today?",
    "hi": "Hi there! Need any help with HR or learning advice?",
    "hey": "Hey! Ready to explore HR info or learning tips?",
    "how are you": "I'm doing great! How can I help you today?",
}

def check_greetings(user_input):
    user_input_lower = user_input.lower()
    for key in greetings:
        if key in user_input_lower:
            return colored(greetings[key], "yellow")
    return None

# -----------------------------
# 7. HR Chatbot Search
# -----------------------------
def hr_search(user_input, threshold=0.30, top_n=2):
    if not model or hr_embeddings is None or len(knowledge_base) == 0:
        return simple_fallback_response(user_input)
        
    user_embedding = model.encode([user_input])
    sims = cosine_similarity(user_embedding, hr_embeddings)[0]
    best_idx = np.argmax(sims)
    score = sims[best_idx]
    
    if score >= threshold:
        return colored(knowledge_base[best_idx] + f" (Confidence: {score:.2f})", "green")
    else:
        top_indices = sims.argsort()[-top_n:][::-1]
        suggestions = [knowledge_base[i] for i in top_indices]
        return colored("I'm not fully confident. Here are some HR suggestions:\n- " + "\n- ".join(suggestions), "green")

# -----------------------------
# 8. Combined Chatbot
# -----------------------------
def professional_chatbot(user_input):

    # 1. Greeting
    greet = check_greetings(user_input)
    if greet:
        return greet

    # 2. HR knowledge FIRST
    hr_answer = hr_search(user_input)
    if hr_answer:
        return hr_answer

    # 3. Learning advice LAST
    if model:
        advice = get_learning_advice(user_input)
        if advice:
            return advice
# -----------------------------
# 9. Interactive Menu Loop
# -----------------------------
if __name__ == "__main__":
    print(colored("="*50, "magenta"))
    print(colored("🤖 HR + Learning Assistant", "magenta"))
    print(colored("="*50, "magenta"))
    
    if model:
        print(colored("✅ AI Mode: Active", "green"))
    else:
        print(colored("⚠️ Simple Mode: Basic responses only", "yellow"))
    
    print(colored("Type 'exit' to quit.\n", "magenta"))
    
    while True:
        try:
            user_text = input(colored("You: ", "blue"))
            if user_text.lower() in ["exit", "quit"]:
                print(colored("Bot: Goodbye! 👋", "magenta"))
                break
            
            if not user_text.strip():
                continue
                
            response = professional_chatbot(user_text)
            print(colored("Bot: ", "green") + str(response))
            
        except KeyboardInterrupt:
            print(colored("\nBot: Goodbye! 👋", "magenta"))
            break
        except Exception as e:
            print(colored(f"Bot: Sorry, an error occurred. Please try again.", "red"))
```

```python
import os
import numpy as np
from sentence_transformers import SentenceTransformer
from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM
from sklearn.metrics.pairwise import cosine_similarity

print("Loading models...")

# --------------------------
# Embedding model
# --------------------------
embed_model = SentenceTransformer("all-MiniLM-L6-v2")

# --------------------------
# Generator model
# --------------------------
model_name = "google/flan-t5-small"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_length=120
)

print("Models loaded!")

# --------------------------
# Load HR documents
# --------------------------
docs_folder = "../hr_docs"
knowledge_base = []

for file in os.listdir(docs_folder):
    if file.endswith(".txt"):
        with open(os.path.join(docs_folder, file), "r", encoding="utf-8") as f:
            text = f.read()
            chunks = text.split("\n\n")
            knowledge_base.extend(chunks)

print("Loaded", len(knowledge_base), "HR paragraphs")

# --------------------------
# Create embeddings
# --------------------------
embeddings = embed_model.encode(knowledge_base)

# --------------------------
# Conversation memory
# --------------------------
chat_history = []

# --------------------------
# RAG Search (Top 3 results)
# --------------------------
def search_docs(query):

    query_emb = embed_model.encode([query])

    sims = cosine_similarity(query_emb, embeddings)[0]

    top_idx = sims.argsort()[-3:][::-1]

    context = "\n".join([knowledge_base[i] for i in top_idx])

    return context

# --------------------------
# Generate Answer
# --------------------------
def generate_answer(question):

    q = question.lower()

    # Greeting handling
    if q in ["hi", "hello", "hii", "hey"]:
        return "Hello! How can I help you with HR today?"

    context = search_docs(question)

    history = "\n".join(chat_history[-6:])

    prompt = f"""
You are a helpful HR assistant.

Conversation history:
{history}

HR Context:
{context}

Employee Question:
{question}

Provide a clear and short answer.
"""
result = generator(prompt, max_length=120)

generated = result[0]["generated_text"]

answer = generated.replace(prompt, "").strip()

    chat_history.append(f"User: {question}")
    chat_history.append(f"Bot: {answer}")

    return answer

# --------------------------
# Chat Loop
# --------------------------
print("\n🤖 HR AI Assistant Ready")
print("Type 'exit' to quit\n")

while True:

    question = input("You: ")

    if question.lower() == "exit":
        break

    answer = generate_answer(question)

    print("Bot:", answer)
```
